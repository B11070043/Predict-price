```python
import pandas as pd
import numpy as np
import tensorflow as tf

import plotly.express as px
import plotly.graph_objects as go


load_model = tf.keras.models.load_model("./Model/model.h5")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/B11070043/Predict-price/main/Dataset/Gold_data.csv?token=GHSAT0AAAAAACH7JK6JTVNIGYWOM4ZEILZYZIPCJRQ", parse_dates=True, index_col='date')
df = df.drop('volume', axis=1)
df
```

```python
df1 = df[2500:-1400]
df2 = df[-1500:]
```

```python
predictions = load_model.predict(df2)
```

```python
# Create the plot
fig = px.line()
fig.add_scatter(x=df1.index, y=df1['open'], name='data open')
fig.add_scatter(x=df2.index, y=df2['open'], name='Actual Open')
fig.add_scatter(x=df2.index, y=predictions[:,0], name='Open')
fig.show()

# Create the plot
fig = px.line()
fig.add_scatter(x=df1.index, y=df1['close'], name='data close')
fig.add_scatter(x=df2.index, y=df2['close'], name='Actual Close')
fig.add_scatter(x=df2.index, y=predictions[:,1], name='Close')

fig.show()
```
